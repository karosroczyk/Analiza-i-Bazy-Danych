```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Zadanie 1
Normal distribution using histograms and cumulative option(sum of each bar).

```python
def count_curve(mu, sigma, bins):
    return 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2))
```

```python
parameter = {}
parameter['N(0,1)'] = (0.0, 1.0)
parameter['N(5,4)'] = (5.0, 4.0) 
samples = {}

for k, (mu, sigma) in parameter.items(): 
    samples[k] = np.random.normal(mu, sigma, 1000)

count, bins, ignored = plt.hist(samples['N(0,1)'], 100, density=True, cumulative=False)
plt.plot(bins, count_curve(parameter['N(0,1)'][0], parameter['N(0,1)'][1], bins), linewidth=2, color='r')
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Normal distribution')
plt.show()
count, bins, ignored = plt.hist(samples['N(0,1)'], 100, density=True, cumulative=True)
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Cumulative distribution function')
plt.show()
```

Zadanie 2 Inverse logit, follows the formula log(x/(1-x))

```python
from scipy.special import logit, expit
y = logit(count)
```

```python
plt.plot(count, y)
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Inverse logit')
plt.show()
```

Zadanie 3 Poisson distribution

```python
data_poisson = []

i = 0
for data in [1, 4, 10]:
    data_poisson.append(np.random.poisson(data, 1000))
    plt.hist(data_poisson[i], 50, density=True)
    plt.grid()
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Poisson distribution')
    i += 1
```

Zadanie 4 Beta distribution

```python
from scipy.stats import beta
```

```python
from collections import namedtuple

Pair = namedtuple("Pair", ["first", "second"])
pairs = [Pair(0.5, 0.5), Pair(5, 1), Pair(1, 3), Pair(2, 2), Pair(2, 5)]
data_beta = []
for pair in pairs:
    data_beta.append(np.linspace(beta.ppf(0, pair.first, pair.second), beta.ppf(1, pair.first, pair.second), 100))
```

```python
i = 0
for (first, second) in pairs:
    plt.plot(data_beta[i], beta.pdf(data_beta[i], first, second))
    plt.grid()
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Beta distribution')  
    i += 1
```
