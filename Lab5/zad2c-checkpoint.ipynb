```python
data = data.fillna(0)
```

```python
data
```

```python
data = data.drop(["year","artist.inverted", "track", "genre", "date.entered", "date.peaked"], axis=1)
```

```python
data.head()
```

```python
df = data.groupby("time")
```

```python
df.describe()
```

```python
dict = {}
for genre, item in df:
    dict[genre] = 0 
```

```python
def final(series, name):
    i = series.between(1,10).sum()
    dict[name] = dict[name] + i;
```

```python
for name, group in df:
    gr1 = group.drop('time', axis=1)
    gr1.apply(lambda item: final(item, name), axis=1)
```

```python
for key, value in dict.items():
   print (key, value)
```

```python
final_data = pandas.DataFrame(dict.items())
```
